```python
import sys
import requests #package for http requests
import bs4 # package for html parsing
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from datetime import date

pd.set_option('display.max_columns', 5000)
```

```python
race = {1993:15,1994:16,1995:17,
1996:16,1997:17,1998:16,1999:16,2000:17,2001:17,2002:17,2003:16,2004:18,2005:19,2006:18,2007:17,2008:18,2009:17,2010:19,
2011:19,2012:18,2013:19,2014:19,2015:20,2016:20,2017:20,2018:21,2019:23}

#generate links

year_range1993 = [year for year in range(date(1993, 1, 1).year, date(1999, 1, 1).year +1)]
year_range2001 = [year for year in range(date(2001, 1, 1).year, date(2010, 1, 1).year +1)]
year_range2013 = [year for year in range(date(2013, 1, 1).year, date(2015, 1, 1).year +1)]
year_range2017 = [year for year in range(date(2017, 1, 1).year, date(2019, 1, 1).year +1)]
```

```python
total_df = pd.DataFrame()

#1993-1999
i=0
x=590
for year in year_range1993:
    x=x+i
    for i in range(race[year]):
        link = 'https://www.formula1.com/en/results.html/'+str(year)+'/races/'+str(x+i)+'/race-result.html'
        slink = 'https://www.formula1.com/en/results.html/'+str(year)+'/races/'+str(x+i)+'/malaysia/starting-grid.html'
        get = requests.get(link)
        soup = bs4.BeautifulSoup(get.text,'lxml')
        table = soup.find_all('table')[0] 
        df = pd.read_html(str(table), flavor='bs4', header=[0])[0]
                
        sget = requests.get(slink)
        ssoup = bs4.BeautifulSoup(sget.text,'lxml')
        stable = ssoup.find_all('table')[0] 
        dfs = pd.read_html(str(stable), flavor='bs4', header=[0])[0]
        
        intermediate = pd.merge(df, dfs[['Driver','Pos']],on='Driver',how='left')
        intermediate['Year']=year
        total_df = total_df.append(intermediate,ignore_index=True)
        
        i+=1

#2000
i=0
x=47
for i in range(race[2000]):
    link = 'https://www.formula1.com/en/results.html/'+str(2000)+'/races/'+str(x)+'/race-result.html'
    slink = 'https://www.formula1.com/en/results.html/'+str(2000)+'/races/'+str(x)+'/malaysia/starting-grid.html'
    
    link_dff = link_dff.append(race_df)
    get = requests.get(link)
    soup = bs4.BeautifulSoup(get.text,'lxml')
    table = soup.find_all('table')[0] 
    df = pd.read_html(str(table), flavor='bs4', header=[0])[0]
                
    sget = requests.get(slink)
    ssoup = bs4.BeautifulSoup(sget.text,'lxml')
    stable = ssoup.find_all('table')[0] 
    dfs = pd.read_html(str(stable), flavor='bs4', header=[0])[0]
        
    intermediate = pd.merge(df, dfs[['Driver','Pos']],on='Driver',how='left')
    intermediate['Year']=year
    total_df = total_df.append(intermediate,ignore_index=True)
    x+=1
```
